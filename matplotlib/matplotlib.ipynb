```python
import matplotlib.pyplot as  plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
plt.plot([1,2,3],[4,5,7])
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,4,6,8,10,12]

plt.plot(x,y, label="2x")
plt.title("NEVER")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend() #it shows the label (2x one)
plt.yticks([0,2,3,4,6,6])# by using ticks we can change the y-axis 
plt.show()
```

```python
#using shorthand notation 
#fmt = '[color][marker][line]'
plt.plot(x,y, 'b.--', label='2x')
plt.legend()
plt.show()
```

```python
gp_1 = np.random.random(50)*100 # 50 values b/w 0 and 100
gp_2 = np.random.random(50)*100
```
